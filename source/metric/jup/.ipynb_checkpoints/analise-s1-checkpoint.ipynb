```python
import numpy as np
from numpy import array
from numpy import corrcoef
import pandas as pd
import matplotlib.pyplot as plt

import os
import sys
```

```python
def measure_correlation(v1,v2):
    return corrcoef(v1,v2)[0,1]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
dataset = pd.DataFrame() 
#Read the csv file
path = (os.path.abspath(".")) + "/../../../dataset/super_trunfo/s1/"
filename = "s1.csv"
file = path + str(filename)
dataset = pd.read_csv(file, delimiter=";", dtype={
    'id': str, 'gp_disp':float, 'titulos':float, 'vitorias':float, 
    'pole_positions':float, 'voltas_rapidas':float
   })
```

```python
dataset
```

```python
dataset.columns
```

```python
dataset.describe()
```

```python
df = dataset
```

```python
columns = ['id']
df.drop(columns, inplace=True, axis=1)
df.head(50)
```

```python
df.describe()
```

```python
scaler = MinMaxScaler() 
scaled_values = scaler.fit_transform(df) 
df.loc[:,:] = scaled_values
```

```python
df
```

```python
print(df.corr())
plt.imshow(df.corr(), cmap=plt.cm.Reds, interpolation='nearest')
plt.colorbar()
tick_marks = [i for i in range(len(df.columns))]
plt.xticks(tick_marks, df.columns, rotation='vertical')
plt.yticks(tick_marks, df.columns)
plt.show()
```
