```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import os
import sys
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
dataset = pd.DataFrame() 
#Read the csv file
path = (os.path.abspath(".")) + "/../data/dsproc/"
filename = "s4.csv"
file = path + str(filename)
dataset = pd.read_csv(file, delimiter=";", dtype={
    'type': str, 'name': str, 'speed':float, 'hp':float, 'kW':float, 
    't100':float, 'cc':float, 'cil':float, 'weight':float
   })
```

```python
dataset.head(10)
```

```python
dataset.columns
```

```python
dataset = dataset.sort_values(by = ['speed'], ascending=False)
dataset.head(10)
```

```python
dataset[['speed', 'hp']]
```

```python
dataset = dataset.sort_values(dataset.columns[3], ascending=False)
dataset.head(10)
```

```python
dataset.describe()
```

```python
df = dataset
```

```python
df.columns
```

```python
columns = ['type', 'name']
df.drop(columns, inplace=True, axis=1)
df.head(5)
```

```python
df.describe()
```

```python
scaler = MinMaxScaler() 
scaled_values = scaler.fit_transform(df) 
df.loc[:,:] = scaled_values
```
